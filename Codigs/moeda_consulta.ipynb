```python
import requests
import pandas as pd
import urllib3
```

```python
urllib3.disable_warnings(urllib3.exceptions.InsecureRequestWarning)
```

```python
url = "https://economia.awesomeapi.com.br/json/last/USD-BRL,EUR-BRL,BTC-BRL,GBP-BRL,ARS-BRL"
response = requests.get(url, verify=False, timeout=10)
dados = response.json()
```

```python
lista_dados = []
nomes = {
    'USDBRL': 'Dólar',
    'EURBRL': 'Euro',
    'BTCBRL': 'Bitcoin',
    'GBPBRL': 'Libra',
    'ARSBRL': 'Peso Argentino'
}
```

```python
lista_dados = []

for codigo, info in dados.items():
    lista_dados.append({
        'Moeda': nomes.get(codigo, codigo),
        'Compra': float(info['bid']),
        'Venda': float(info['ask']),
        'Variacao': float(info['pctChange']),
        'Maxima': float(info['high']),
        'Minima': float(info['low'])
    })

```

```python
df = pd.DataFrame(lista_dados)
```

```python
print("Colunas disponíveis:", df.columns.tolist())
print("\n")
```

```python
df_formatado = df.copy()

df_formatado['Compra'] = df['Compra'].apply(lambda x: f'R$ {x:,.2f}')
df_formatado['Venda'] = df['Venda'].apply(lambda x: f'R$ {x:,.2f}')
df_formatado['Variacao'] = df['Variacao'].apply(lambda x: f'{x:+.2f}%')
df_formatado['Maxima'] = df['Maxima'].apply(lambda x: f'R$ {x:,.2f}')
df_formatado['Minima'] = df['Minima'].apply(lambda x: f'R$ {x:,.2f}')
```
