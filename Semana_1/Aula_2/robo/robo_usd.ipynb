```python
import requests
import urllib3

urllib3.disable_warnings(urllib3.exceptions.InsecureRequestWarning)

# --- CONFIGURAÇÕES (PARÂMETROS DO ROBÔ) ---
MOEDA_ORIGEM = "USD"
MOEDA_DESTINO = "BRL"
LIMITE_ALERTA = 6.00  # Nosso gatilho de negócio

print("Iniciando o Robô de Monitoramento...")

# 1. BUSCAR A COTAÇÃO (GET)
# Acessando a URL da API da AwesomeAPI
url = f"https://economia.awesomeapi.com.br/last/{MOEDA_ORIGEM}-{MOEDA_DESTINO}"
resposta = requests.get(url, timeout=10, verify=False)

# 2. TRATAR OS DADOS (A "Tradução")
if resposta.status_code == 200: # 200 significa "OK, consegui os dados"
    dados = resposta.json() # Transforma o texto da internet em um Dicionário Python
    
    # Navegando no Dicionário: USDBRL -> bid (preço de compra)
    cotacao_atual = float(dados['USDBRL']['bid'])
    
    print(f"Cotação capturada: R$ {cotacao_atual:.2f}")

    # 3. TOMADA DE DECISÃO (Regra de Negócio)
    if cotacao_atual > LIMITE_ALERTA:
        print("ALERTA: Dólar acima do limite! Preparando e-mail...")
        
        # Montando o E-mail
        print(f"URGENTE: Dólar a R$ {cotacao_atual:.2f}")
        
        print(f"""
        Olá,
        
        O sistema de monitoramento detectou uma alta no câmbio.
        Valor Atual: R$ {cotacao_atual:.2f}
        Limite Definido: R$ {LIMITE_ALERTA:.2f}
        
        Recomendação: Segurar a compra.
        
        Att,
        Robô Python
        """)
    
    else:
        print("Cotação dentro do esperado. Nenhuma ação necessária.")

else:
    print("Erro ao conectar na API. Verifique a internet.")
```

```python
import requests  # O carteiro que busca dados na internet
import time


# --- CONFIGURAÇÕES (PARÂMETROS DO ROBÔ) ---
MOEDA_ORIGEM = "USD"
MOEDA_DESTINO = "BRL"
LIMITE_ALERTA_MAX = 6.00  # Nosso gatilho de negócio
LIMIT_ALERTA_MIN = 4.00
print("Iniciando o Robô de Monitoramento...")

cotacao_atual = 3.0
## Repete para sempre ou caso de erro
while True: # cotacao_atual < LIMITE_MAX and cotacao_atual > LIMITE_MIN
    # 1. BUSCAR A COTAÇÃO (GET)
    # Estamos acessando a URL da API da AwesomeAPI
    url = f"https://economia.awesomeapi.com.br/last/{MOEDA_ORIGEM}-{MOEDA_DESTINO}"
    resposta = requests.get(url, verify=False)
    
    # 2. TRATAR OS DADOS (A "Tradução")
    if resposta.status_code == 200: # 200 significa "OK, consegui os dados"
        dados = resposta.json() # Transforma o texto da internet em um Dicionário Python
        
        # Navegando no Dicionário: USDBRL -> bid (preço de compra)
        cotacao_atual = float(dados['USDBRL']['bid'])
        
        print(f"Cotação capturada: R$ {cotacao_atual:.2f}")
    
        # 3. TOMADA DE DECISÃO (Regra de Negócio)
        if cotacao_atual > LIMITE_ALERTA_MAX:
            print("ALERTA: Dólar acima do limite! Preparando e-mail...")
            
            # Montando o E-mail
            print(f"URGENTE: Dólar a R$ {cotacao_atual:.2f}")
            
            print(f"""
            Olá,
            
            O sistema de monitoramento detectou uma alta no câmbio.
            Valor Atual: R$ {cotacao_atual:.2f}
            Limite Definido: R$ {LIMITE_ALERTA_MAX:.2f}
            
            Recomendação: Segurar a compra.
            
            Att,
            Robô Python
            """)

            break
        elif cotacao_atual < LIMIT_ALERTA_MIN:
            print("ALERTA: Dólar abaixo do limite! Preparando e-mail...")
            
            # Montando o E-mail
            print(f"URGENTE: Dólar a R$ {cotacao_atual:.2f}")
            
            print(f"""
            Olá,
            
            O sistema de monitoramento detectou uma baixa no câmbio.
            Valor Atual: R$ {cotacao_atual:.2f}
            Limite Definido: R$ {LIMITE_ALERTA:.2f}
            
            Recomendação: Segurar a compra.
            
            Att,
            Robô Python
            """)

            break
        else:
            print("Cotação dentro do esperado. Nenhuma ação necessária.")
    
    else:
        print("Erro ao conectar na API. Verifique a internet.")
        break # Comando que termina repetição forçadamente
        
    time.sleep(10)
```

```python
import requests
import time
from datetime import datetime

MOEDA_ORIGEM = "USD"
MOEDA_DESTINO = "BRL"
LIMITE_ALERTA = 6.00
ARQUIVO_SAIDA = "cotacoes.txt"

print("Iniciando o Robô de Monitoramento...")

while True:
    url = f"https://economia.awesomeapi.com.br/last/{MOEDA_ORIGEM}-{MOEDA_DESTINO}"
    
    try:
        resposta = requests.get(url)
        
        if resposta.status_code == 200:
            dados = resposta.json()
            cotacao_atual = float(dados['USDBRL']['bid'])
            
            print(f"Cotação capturada: R$ {cotacao_atual:.2f}")
            
            # --- SALVANDO EM ARQUIVO DE TEXTO (SIMPLES) ---
            # Montamos uma linha de texto com Data - Valor
            data_hora = datetime.now().strftime('%d/%m/%Y %H:%M:%S')
            linha = f"{data_hora} | {MOEDA_ORIGEM}->{MOEDA_DESTINO} | R$ {cotacao_atual:.2f}\n"
            
            # 'a' significa append (adicionar ao final sem apagar o que já existe)
            with open(ARQUIVO_SAIDA, 'a') as arquivo:
                arquivo.write(linha)
            
            print(f"-> Salvo no arquivo {ARQUIVO_SAIDA}")
            
            # Regra de Negócio (Alerta)
            if cotacao_atual > LIMITE_ALERTA:
                print("ALERTA: Cotação acima do limite!")
        else:
            print("Erro ao conectar na API.")
            
    except Exception as e:
        print(f"Erro: {e}")
        
    time.sleep(10)
```
