```python
aposentados['CPF'] = aposentados['CPF'].astype(str).str.replace(r'[^\d]', '', regex=True).str.strip()
devedores['CPF/CNPJ DEVEDOR'] = devedores['CPF/CNPJ DEVEDOR'].astype(str).str.replace(r'[^\d]', '', regex=True).str.strip()


resultado = pd.merge(
    aposentados,
    devedores,
    left_on='CPF',
    right_on='CPF/CNPJ DEVEDOR',
    how='inner'
)


print(f"Total de aposentados: {len(aposentados)}")

resultado
```
