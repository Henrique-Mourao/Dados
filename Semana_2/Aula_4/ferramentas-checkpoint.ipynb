### Combinando dados

```python
# DataFrame de Funcionários
dados_func = {
    'id_func': [1, 2, 3, 4, 5],
    'nome': ['Ana', 'Carlos', 'Beatriz', 'Davi', 'Eva'],
    'id_dept': [101, 102, 101, 103, 102]  # Note que os códigos se repetem
}
df_func = pd.DataFrame(dados_func)

# DataFrame de Departamentos
dados_depts = {
    'id_dept': [101, 102, 103],
    'nome_dept': ['Recursos Humanos', 'TI', 'Vendas']
}
df_depts = pd.DataFrame(dados_depts)
```

```python
df_completo = pd.merge(df_func, df_depts)
df_completo
```

```python
df_completo = pd.merge(df_func, df_depts, left_on='id_dept', right_on='id_dept')
df_completo
```

```python
dados_func = {
    'id_func': [6, 7, 8, 9, 10],
    'nome': ['Patrick', 'Jaqueline', 'Bianca', 'Raquel', 'Andrei'],
    'id_dept': [101, 102, 101, 103, 102]
}

df_func_novos = pd.DataFrame(dados_func)
```

```python
df_todos = pd.concat([df_func, df_func_novos], ignore_index=True)
df_todos
```
