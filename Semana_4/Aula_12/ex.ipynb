```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Gerando notas entre 5 e 10
notas = np.random.randint(0, 11, 50)

# Inserindo um valor fora da curva (outlier)
notas[15] = 6

df_notas = pd.DataFrame({
    'Notas': notas
})

df_notas
```

```python
media = df_notas['Notas'].mean()
mediana = df_notas['Notas'].median()

print(f"Notas Média: R$ {media:.2f}")
print(f"Notas Mediana: R$ {mediana:.2f}")
```

```python
fig, ax = plt.subplots(figsize=(10, 5))

ax.plot(df_notas['Notas'], 
    color='#2E86C1',
    linewidth=2,
    marker='o',         # Bolinha em cada ponto
)

ax.set_title('Variação das notas de atendimentos', fontsize = 14)
ax.set_xlabel('Notas')
ax.axhline(media, color='green', linestyle='--', linewidth=1.5, label=f'Média = {media:.2f}')
ax.axhline(mediana, color='orange', linestyle=':', linewidth=1.5, label=f'Mediana = {mediana:.2f}')
ax.legend()
```
