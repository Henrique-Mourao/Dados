```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python

sns.set_theme(style="whitegrid")

tips = pd.read_csv('tips.csv')

tips.head()
```

```python
print("\n=== Estatísticas Básicas ===")
print(tips.describe())
print("--------------------------")
 
```

```python
#criando a figura(sim, ainda usando plt.subplots para controlar o tamanho!)
fig, ax = plt.subplots(figsize = (10,6))
 
#o comando mágico do seaborn
sns.histplot(data=tips, x='total_bill', kde=False, color='teal', ax=ax)
 
ax.set_title('Distribuição do Valor das Contas (Histograma)', fontsize=14)
ax.set_xlabel('Valor da Conta ($)')
ax.set_ylabel('Frequência (Qtd Mesas)')
 
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(10, 6))

sns.scatterplot(data=tips, x='total_bill', y='tip', s=100, alpha=0.7, color='teal', ax=ax)
ax.set_title('Conta vs Gorjeta')

plt.show()
```

```python
fig, ax = plt.subplots(figsize=(10, 6))

sns.regplot(
    data=tips, 
    x='total_bill', 
    y='tip', 
    scatter_kws={'s': 200, 'alpha': 0.7},  
    line_kws={'color': 'blue', 'linewidth': 2}, 
    color='red',
    ax=ax
)

plt.show()
```
