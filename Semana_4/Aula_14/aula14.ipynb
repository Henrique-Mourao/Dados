```python
import seaborn as sns
import matplotlib.pyplot as plt   
import pandas as pd
import matplotlib.image as mpimg
```

```python

sns.set_theme(style="whitegrid")

geo = pd.read_csv('df_geo.csv')

geo.head()
```

```python
fig, sx = plt.subplots(figsize=(10, 8))

sns.scatterplot(data=geo,
x='Longitude',
y='Latitude',
)

plt.show()
```

```python
fig, ax = plt.subplots(figsize=(12, 8))

sns.scatterplot(
    data=geo,
    x='Longitude',
    y='Latitude',
    hue='Regiao',
    size='Faturamento',
    sizes=(50, 400),
    palette='viridis',
    alpha=0.7,
    ax=ax,
    edgecolor='black'
)

ax.set_title('Mapa de Lojas: Distribuio Geogrfica', fontsize=15, fontweight='bold')
ax.set_xlabel('Longitude')
ax.set_ylabel('Latitude')
ax.set_aspect('equal', adjustable='datalim')

ax.legend(bbox_to_anchor=(1.05, 1), loc='upper left', borderaxespad=0)

sns.despine(left=True, bottom=True, top=True, right=True)

plt.tight_layout()
plt.show()
```
