```python
import sqlite3
import pandas as pd 

DB_PATH = 'escola_v2.db' # Caminho para o banco de dados SQLite
```

```python
def executar_query(sql_query):
    try:
        conn = sqlite3.connect(DB_PATH)
        df_resultado = pd.read_sql_query(sql_query, conn)
        conn.close()
        return df_resultado
    except:
        print("Erro")
```

```python
query_alunos_cidade = 'SELECT nome, cidade FROM alunos WHERE cidade = "São Paulo"'

executar_query(query_alunos_cidade)
```

```python
query_alunos_comecam_C = 'SELECT nome FROM alunos WHERE nome LIKE "C%"'

executar_query(query_alunos_comecam_C)
```

```python
query_cursos_valor = 'SELECT * FROM cursos WHERE preco_mensal BETWEEN 300.0 AND 500.0'

executar_query(query_cursos_valor)
```

```python
query_matriculas_ativas_ou_concluidas = 'SELECT * FROM matriculas WHERE status IN ("Ativo", "Concluido")'

query_matriculas_ativas_ou_concluidas_count = '''
SELECT
    status, 
    count (status) AS qtd 
FROM
    matriculas 
WHERE
    status IN ("Ativo", "Concluido") 
GROUP BY 1
'''

executar_query(query_matriculas_ativas_ou_concluidas_count)
```
