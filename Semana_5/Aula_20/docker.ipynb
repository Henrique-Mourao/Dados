```python
import pandas as pd 
from sqlalchemy import create_engine
```

```python
connection_string = 'postgresql://postgres:senha123@localhost:5432/postgres'

try:
    engine = create_engine(connection_string)
    print("✓ Conectado ao Docker com sucesso!")
    
    # 2. Criando DataFrame de teste
    df_teste = pd.DataFrame({
        'id': [1, 2, 3],
        'nome': ['Ana', 'Bruno', 'Carla'],
        'cargo': ['Analista', 'Cientista de Dados', 'Engenheira de Dados']
    })
    
    print("\nDados a serem inseridos:")
    print(df_teste)
    
    # 3. Escrevendo dados no Docker
    print("\nEscrevendo dados no Docker...")
    df_teste.to_sql('funcionarios',
                    engine,
                    if_exists='replace',
                    index=False)
    print("✓ Dados escritos com sucesso!")
    
    # 4. Lendo dados do Docker
    print("\nLendo dados do Docker...")
    df_lido = pd.read_sql('SELECT * FROM funcionarios', engine)
    print("✓ Dados lidos com sucesso!")
    print("\n" + "="*50)
    print(df_lido)
    print("="*50)
    
except Exception as e:
    print(f"✗ Erro: {e}")
```
