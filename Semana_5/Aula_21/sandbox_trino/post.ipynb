```python
import pandas as pd
from sqlalchemy import create_engine

# Conectando direto no Postgres para criar os dados
engine_pg = create_engine('postgresql://postgres:senha123@localhost:5432/postgres')

df = pd.DataFrame({
    'id': range(1, 6),
    'produto': ['Notebook', 'Mouse', 'Teclado', 'Monitor', 'Cabo HDMI'],
    'preco': [3500, 50, 150, 1200, 30]
})

df.to_sql('produtos_loja', engine_pg, if_exists='replace', index=False)
df
```

```python
df = pd.DataFrame({
    'id': range(1, 6),
    'nome': ['gaule', 'pedro', 'thiago', 'joão', 'julio'],
    'cargo': ['ceo', 'analista', 'gerente', 'diretor', 'estagiário']
})

df.to_sql('funcionarios', engine_pg, if_exists='replace', index=False)
df
```

```python
query = """
SELECT *
FROM produtos_loja
WHERE preco > 100;
"""

resultado = pd.read_sql(query, engine_pg)

print(resultado)
```

```python
query = "SELECT produto, preco FROM produtos_loja;"
df_select = pd.read_sql(query, engine_pg)
```

```python
query = "SELECT * FROM produtos_loja ORDER BY preco DESC;"
df_sorted = pd.read_sql(query, engine_pg)
```

```python
query = "SELECT * FROM produtos_loja ORDER BY preco ASC;"
df_ascending = pd.read_sql(query, engine_pg)
```
