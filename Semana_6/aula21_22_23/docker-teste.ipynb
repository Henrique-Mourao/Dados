```python
import pandas as pd
from sqlalchemy import create_engine

#1- conectando
connection_string = 'postgresql://postgres:senha123@'\
'localhost:5432/postgres'
engine = create_engine(connection_string)
```

```python
#2- criando um DataFrame
df_teste = pd.DataFrame({
    'id' : [1, 2, 3, 5],
    'nome' : ['Ana', 'Gustavo', 'Henrique', 'Patrik'],
    'cargo' : ['CEO', 'Analista', 'Gerente', 'PO']
})
df_teste
```

```python
#3-escrevendo no banco
print('Escrevendo dados no Docker...')
df_teste.to_sql('funcionarios',
                engine,
                if_exists='replace', #adiciona, mesmo se  tiver iguais e nao repete
                index=False)
```

```python
#4- select
print('Lendo dados do Docker...')
df_lido = pd.read_sql(
    'SELECT * FROM funcionarios',
    engine
    )
df_lido
```
