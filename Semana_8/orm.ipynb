```python
from sqlalchemy import create_engine, String
from sqlalchemy.orm import \
    DeclarativeBase, ForeignKey,\
    Mapped, mapped_column
```

```python


from email.mime import base


STRING_DB = ""


class Base(DeclarativeBase):
    pass
    
class Livro(Base):
    __tablename__ = "livros"

    id: Mapped[int] = mapped_column(primary_key=True)
    nome: Mapped[str] = mapped_column(String(100))
    autor: Mapped[str] = mapped_column(String(100))
    preco: Mapped[float] = mapped_column()
    quantidade: Mapped[int] = mapped_column()
    editora: Mapped[str] = mapped_column(String(100))
    isbn: Mapped[str] = mapped_column(String(20))
    sinopse: Mapped[str] = mapped_column(String(500))
    genero: Mapped[int] = mapped_column(ForeignKey("categorias.id"))
    
class Categoria(Base):
    __tablename__ = "categorias"

    id: Mapped[int] = mapped_column(primary_key=True)
    nome: Mapped[str] = mapped_column(String(100))
    
class Usuario(Base):
    __tablename__ = "usuarios"

    id: Mapped[int] = mapped_column(primary_key=True)
    nome: Mapped[str] = mapped_column(String(100))
    email: Mapped[str] = mapped_column(String(100))
    senha: Mapped[str] = mapped_column(String(100))
    endereco: Mapped[str] = mapped_column(String(200))
    telefone: Mapped[str] = mapped_column(String(20))

class Avaliacao(Base):
    __tablename__ = "avaliacoes"

    id: Mapped[int] = mapped_column(primary_key=True)
    usuario_id: Mapped[int] = mapped_column(ForeignKey("usuarios.id"))
    livro_id: Mapped[int] = mapped_column(ForeignKey("livros.id"))
    nota: Mapped[int] = mapped_column()
    comentario: Mapped[str] = mapped_column(String(500))
    
class Multa(Base):
    __tablename__ = "multas"

    id: Mapped[int] = mapped_column(primary_key=True)
    usuario_id: Mapped[int] = mapped_column(ForeignKey("usuarios.id"))
    valor: Mapped[float] = mapped_column()
    descricao: Mapped[str] = mapped_column(String(500))

class Venda(Base):
    __tablename__ = "vendas"

    id: Mapped[int] = mapped_column(primary_key=True)
    usuario_id: Mapped[int] = mapped_column(ForeignKey("usuarios.id"))
    livro_id: Mapped[int] = mapped_column(ForeignKey("livros.id"))
    data_venda: Mapped[str] = mapped_column(String(20))
    quantidade: Mapped[int] = mapped_column()
    valor_total: Mapped[float] = mapped_column()

class Emprestimo(base):
    __tablename__ = "emprestimos"

    id: Mapped[int] = mapped_column(primary_key=True)
    usuario_id: Mapped[int] = mapped_column(ForeignKey("usuarios.id"))
    livro_id: Mapped[int] = mapped_column(ForeignKey("livros.id"))
    data_emprestimo: Mapped[str] = mapped_column(String(20))
    data_devolucao: Mapped[str] = mapped_column(String(20))

engine = create_engine(STRING_DB)
Base.metadata.create_all(engine)
```
